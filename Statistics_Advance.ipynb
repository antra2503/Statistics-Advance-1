# Theory Questions

**Q1. What is a random variable in probability theory?**

--> A random variable is a function that assigns numerical values to the outcomes of a random experiment. It is used to quantify uncertainty and can take on different values based on probability. There are two types of random variables: discrete (having a countable number of values) and continuous (having an uncountable range of values).

**Q2. What are the types of random variables?**

--> Random variables are classified into two types:

Discrete Random Variable: Takes only specific values, such as whole numbers. Example: The number of heads in 10 coin flips.

Continuous Random Variable: Can take any value within a range. Example: The height of students in a class.

**Q3. What is the difference between discrete and continuous distributions?**

--> Discrete Distribution: Deals with discrete random variables and is defined using a probability mass function (PMF). Example: Binomial distribution.

Continuous Distribution: Deals with continuous random variables and is defined using a probability density function (PDF). Example: Normal distribution.

**Q4. What are probability distribution functions (PDF)?**

--> A probability distribution function describes how probabilities are distributed over the values of a random variable. It defines the likelihood of different outcomes occurring in a statistical experiment.

**Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

--> PDF (Probability Density Function): Shows the likelihood of a specific value occurring in a continuous distribution.

CDF (Cumulative Distribution Function): Represents the probability that a random variable takes a value less than or equal to a given number. The CDF always increases and reaches 1 at the maximum value.

**Q6. What is a discrete uniform distribution?**

--> A discrete uniform distribution is one where all possible outcomes have the same probability. For example, rolling a fair six-sided die results in equal probability (1/6) for each face (1 to 6).

**Q7. What are the key properties of a Bernoulli distribution?**

--> It represents a single trial with two possible outcomes: success (1) and failure (0).

The probability of success is p, and the probability of failure is 1 - p.

The mean is p, and the variance is p(1 - p).

**Q11. What are the characteristics of a normal distribution?**

--> Symmetric bell-shaped curve.

Defined by mean μ and standard deviation σ.

Most values fall within three standard deviations from the mean.

**Q12. What is the standard normal distribution, and why is it important?**

--> A normal distribution with mean = 0 and standard deviation = 1.

It allows us to standardize data and compare different datasets.

**Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

--> The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population distribution. It is crucial because it justifies the use of normal approximations in statistical inference.

**Q14. How does the Central Limit Theorem relate to the normal distribution?**

--> The CLT explains why many real-world distributions appear normal even if the original data is not. It shows that averaging independent observations tends to produce a normal shape.

**Q15. What is the application of Z statistics in hypothesis testing?**

--> Z-statistics are used when testing hypotheses for large samples with a known standard deviation. It helps compare observed values with expected values under the null hypothesis.

**Q17. What are point estimates and interval estimates in statistics?**

--> Point estimate: A single value used to estimate a population parameter (e.g., sample mean).

Interval estimate: A range of values within which the parameter is expected to lie (e.g., confidence interval).

**Q18. What is the significance of confidence intervals in statistical analysis?**

--> A confidence interval provides a range of values that likely contain the population parameter. For example, a 95% confidence interval means that 95% of such intervals would contain the true mean.

**Q19. What is the relationship between a Z-score and a confidence interval?**

--> A confidence interval is constructed using a Z-score. For a 95% confidence interval, we use a Z-score of 1.96 to calculate the range around the mean.

**Q20. How are Z-scores used to compare different distributions?**

--> Since Z-scores standardize data, they allow us to compare different distributions, even if they have different units and scales.

**Q21. What are the assumptions for applying the Central Limit Theorem?**

--> The sample size should be large (n > 30).

The samples must be independent.

The population should not be highly skewed.

```python
#1. Generate a random variable and display its value

import numpy as np

random_variable = np.random.randint(1, 101)
print("Random Variable:", random_variable)
```

```python
#2. Generate a discrete uniform distribution and plot the PMF

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint

low, high = 1, 6
samples = randint.rvs(low, high + 1, size=10000)

sns.histplot(samples, bins=6, discrete=True, stat="probability", kde=False)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution (Rolling a Die)")
plt.show()
```

```python
#3. Calculate the PDF of a Bernoulli distribution

from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

p = 0.5
for x in [0, 1]:
    print(f"P(X={x}) = {bernoulli_pdf(p, x)}")
```

```python
#4. Simulate a binomial distribution and plot its histogram

n, p = 10, 0.5
binomial_samples = np.random.binomial(n, p, 1000)

plt.hist(binomial_samples, bins=11, density=True, alpha=0.6, color='b', edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
#5. Create a Poisson distribution and visualize it

from scipy.stats import poisson

lambda_ = 5
samples = poisson.rvs(lambda_, size=1000)

sns.histplot(samples, bins=15, discrete=True, kde=False)
plt.xlabel("Events")
plt.ylabel("Frequency")
plt.title("Poisson Distribution (λ=5)")
plt.show()
```

```python
#6. Calculate and plot the CDF of a discrete uniform distribution

import scipy.stats as stats

x = np.arange(1, 7)
cdf_values = stats.randint.cdf(x, 1, 7)

plt.step(x, cdf_values, where="mid", color='red', label="CDF")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of a Discrete Uniform Distribution")
plt.legend()
plt.show()
```

```python
#7. Generate and visualize a continuous uniform distribution

samples = np.random.uniform(0, 10, 1000)

plt.hist(samples, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Continuous Uniform Distribution (0,10)")
plt.show()
```

```python
#8. Simulate data from a normal distribution and plot its histogram

mean, std_dev = 0, 1
samples = np.random.normal(mean, std_dev, 1000)

sns.histplot(samples, bins=30, kde=True)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Normal Distribution")
plt.show()
```

```python
#9. Calculate and plot Z-scores from a dataset

from scipy.stats import zscore

data = np.random.normal(50, 10, 1000)
z_scores = zscore(data)

sns.histplot(z_scores, bins=30, kde=True)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Z-score Distribution")
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python

sample_means = [np.mean(np.random.exponential(1, 30)) for _ in range(1000)]

sns.histplot(sample_means, bins=30, kde=True)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem in Action")
plt.show()
```

```python
#11. Simulate multiple samples from a normal distribution and verify CLT

sample_sizes = [5, 10, 30, 50]
plt.figure(figsize=(10, 6))

for size in sample_sizes:
    sample_means = [np.mean(np.random.normal(0, 1, size)) for _ in range(1000)]
    sns.kdeplot(sample_means, label=f"Sample size {size}")

plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem Verification")
plt.legend()
plt.show()
```

```python
#14. Calculate a Z-score for a given data point

X = 75
mean, std = 50, 10
z = (X - mean) / std
print("Z-score:", z)
```

```python
#15. Implement hypothesis testing using Z-statistics

from scipy.stats import norm

sample_mean = 55
population_mean = 50
std_dev = 10
n = 30

z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z_score)

print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
#16. Create a confidence interval

import scipy.stats as stats

data = np.random.normal(50, 10, 100)
confidence = 0.95
mean = np.mean(data)
std_err = stats.sem(data)
margin = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * std_err

lower, upper = mean - margin, mean + margin
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#21. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np
import scipy.stats as stats

data = np.random.normal(50, 10, 1000)

# Range
data_range = np.max(data) - np.min(data)

# Interquartile Range (IQR)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")
```
